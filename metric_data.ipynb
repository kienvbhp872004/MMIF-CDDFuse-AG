```python
import pandas as pd
import matplotlib.pyplot as plt

df_ct_mri = pd.read_csv("fusion_metrics_results_ct_mri.csv")
df_mri_spect = pd.read_csv("fusion_metrics_results_mri_spect.csv")

# lấy 10 dòng đầu và làm tròn
df_ct_mri = df_ct_mri.head(10).round(3)
df_mri_spect = df_mri_spect.head(10).round(3)

def save_table(df, title, filename):
    fig, ax = plt.subplots(figsize=(14,5))
    ax.axis('off')

    table = ax.table(
        cellText=df.values,
        colLabels=df.columns,
        loc='center'
    )

    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1,1.5)

    plt.title(title, fontsize=14, pad=20)
    plt.savefig(filename, bbox_inches='tight', dpi=300)
    plt.close()

# lưu ảnh
save_table(df_ct_mri, "Fusion Metrics CT-MRI", "ct_mri_metrics.png")
save_table(df_mri_spect, "Fusion Metrics MRI-SPECT", "mri_spect_metrics.png")
```
